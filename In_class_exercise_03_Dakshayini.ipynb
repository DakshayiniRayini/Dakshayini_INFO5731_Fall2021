Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):
import nltk

from bs4 import BeautifulSoup as bs
import pandas as p
import requests as reqs
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud
from nltk.corpus import stopwords

#collecting sample text data from IMDb "KGF2 movie reviews"

url = 'https://www.imdb.com/title/tt10698680/reviews?ref_=tt_ov_rt'

#Storing the reviews into a dataframe
dataframe = p.DataFrame(columns=['Review'])
request = reqs.get(url).text
soup = bs(request,'html.parser')

#collecting reviews as text
r_text = soup.find_all('div', attrs={'class' : 'text show-more__control'})

#taking a empty list initially
r_list = []
for i in range(len(r_text)):
    r_list.append(r_text[i].get_text())

#assigining list of reviews to the data frame    
dataframe['Review'] = r_list

dataframe['word_count'] = dataframe['Review'].apply(lambda x: len(str(x).split(" ")))
dataframe[['Review','word_count']].head()

```

```python
#character count

dataframe['char_count'] = dataframe['Review'].str.len() ## including spaces
dataframe[['Review','char_count']].head()
```

```python
# Getting Stopwords

from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

dataframe['stopwords'] = dataframe['Review'].apply(lambda x: len([x for x in x.split() if x in stop]))
dataframe[['Review','stopwords']].head()
```

```python
#hashtags in the reviews
dataframe['hastags'] = dataframe['Review'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
dataframe[['Review','hastags']].head()
```

```python
#numerics in the reviews
dataframe['numerics'] = dataframe['Review'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
dataframe[['Review','numerics']].head()
```

```python
#Upper characters in the reviews
dataframe['upper'] = dataframe['Review'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
dataframe[['Review','upper']].head()
```
